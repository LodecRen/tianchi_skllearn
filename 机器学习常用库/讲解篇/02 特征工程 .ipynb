# 序言
因为教程中数据太多,V0-V37,target 总共38个特征,一个标签  
这个讲解就是为了熟悉理解其中的库和函数方法的使用  
* 生成一个 两个特征, 一个target的数据

* 例如: code1: train_data = train_data[train_data['V9']>-7.5]

在执行code1之前: count = 2888  
在执行code1之后: count = 2886  
在删除 V9的两个异常值的同时,也将对应行其他特征的值也删除了,保持的数据的对齐  
下面写了一个示例demo:
```python
import pandas as pd

# 给定的 train_data['V9'] 值
V0_values = [-10.0, -8.57142857, -7.14285714, -5.71428571, -4.28571429,
             -2.85714286, -1.42857143, 0.0, 1.42857143, 2.85714286,
             4.28571429, 5.71428571, 7.14285714, 8.57142857, 10.0]
V1_values = [-10.0, -8.57142857, -7.14285714, -5.71428571, -4.28571429,
             -2.85714286, -1.42857143, 0.0, 1.42857143, 2.85714286,
             4.28571429, 5.71428571, 7.14285714, 8.57142857, 10.0]

# 创建 DataFrame
train_data = pd.DataFrame({'V0': V0_values, 'V1': V1_values})

print("原始train_data: ")
print(train_data)

# 删除异常值
train_data = train_data[train_data['V0'] > -7.5]

# 输出新的 train_data值
print("\n删除异常值后的train_data: ")
print(train_data)
```
