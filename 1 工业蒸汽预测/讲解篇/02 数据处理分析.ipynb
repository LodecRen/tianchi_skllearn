### 1.2 双变量分析
变量之间的关系:  
* 连续型和连续型
* 离散型和连续型
* 离散型和离散型

(1)连续型和连续型
* 绘制散点图: 观察变量之间的关系是线性还是非线性;
* 计算相关性(correlation): 反应数据间关系的强弱; $cor= \frac{Cov(x,y)}{\sqrt{Var(x)*Var(y)}}$
    * 检查特征相关性的意义, 举个例子(可能不太适合,理解为主): 比如说利用三十个特征来判断一个人的性别, 其中一个特征是年龄, 那么年龄这个数据其实是对于判断性别无意义的;应当剔除;

绝对值后: 
* 0-0.09 无相关性
* 0.1-0.3 弱相关
* 0.3-0.5 中等相关
* 0.5-1 强相关




(2)离散型和离散型
* 双向表
* 柱状堆叠图
* 卡方检测

双向表: 利用双向表建立频率和频次的占比;  
* 例如:一个学生在不同科目中的成绩表格



柱状堆叠图: 比双向表更直观
* 假设我们有一个小型餐厅，我们想比较每个月不同类型食物的销售情况

   

卡方检测: 主要用于两个和两个以上样本率(构成比)及两个二值型离散变量的关联性分析,即比较理论频次与实际频次的吻合程度或拟合优度







(2)离散型和连续型
* 小提琴图()

可以分析类别变量在不同类别时,另一个连续变量的分布情况。如图1-2-4所示,通过绘制小提琴图,可以对比在类别变量为low,medium,high三个不同类别时,连续变量price的分布情况







```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 生成示例数据集
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],  # 添加Age数据
    'Smoker': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']
}

# 将数据转换为DataFrame格式
import pandas as pd
df = pd.DataFrame(data)

# 绘制小提琴图
plt.figure(figsize=(8, 6))
sns.violinplot(x='Gender', y='Age', data=df, palette='Set2')  # 修改y为数值型变量Age
plt.title('Violin Plot of Gender vs. Age')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()

# 进行卡方检验
crosstab = pd.crosstab(df['Gender'], df['Smoker'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print(f'Chi-square statistic: {chi2}')
print(f'p-value: {p}')
```
