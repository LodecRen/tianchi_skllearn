## 1.2 泛化和正则化  

### 泛化能力  
1. 模型不是越复杂越好, 复杂度模型在训练集上的准确率甚至能达到 100%, 仅仅针对这一个训练集;
    1. 这就是为什么左图, 测试集的准确率先升后降, 随着模型越来越贴合训练集, 它对未知数据的预测就越不准确;
2. 最好的模型就是在测试集上准确率最高的模型;

正则化就是在损失函数后, 加一个正则项, 正则项就是用来减少非主要特征的影响;

## 1.3 回归模型评估指标
MSE 最常用, MAE计算机求绝对值不太方便,不常用; 
* 平均绝对值误差 (MAE)
* 均方误差 (MSE) 
* 均方根误差 (RMSE)
* R 平方值

### 平均误差
MAE, MSE, RMSE都是基于平均误差, 其值越接近0, 准确率越高;

### R 平方值
$R^2\in[0,1]$ 评价指标是接近1为好. 但是$R^2$ 有局限性, 很多时候不是 $R^2$ 值越高越好. 

## 分类模型评价指标：

* 准确率（Accuracy）：预测正确的样本数占总样本数的比例，是最常用的分类模型评价指标之一
* 精确率（Precision）：在所有被分类为正例的样本中，实际为正例的比例，用于衡量分类器的预测准确性
* 召回率（Recall）：在所有实际为正例的样本中，被正确分类为正例的比例，用于衡量分类器对正例的识别能力
* F1-score：精确率和召回率的调和平均值，综合考虑了两者的性能
* ROC曲线和AUC值：ROC曲线是以假正例率（False Positive Rate，FPR）为横轴，真正例率（True Positive Rate，TPR）为纵轴的曲线，AUC值表示ROC曲线下的面积，用于衡量分类器的性能

## 1.5 交叉验证
特别常用的机器学习评估方法, 就是将原始数据分为训练集和验证集.
* 简单交叉验证
* K折交叉验证
* 留1交叉验证
* 留P交叉验证

### 简单/K折
简单其实就是K=2的K折, 将数据分为训练, 测试两个部分, 一般是73开.  

K折: K大于3, 将原始数据分为K份, 一般是均等分.   
每个子集分别做一次验证集, 另外的K-1份做训练集, 训练K次, 取准确率的平均值.

### 留P交叉验证
将数据集中的P个样本, 拆分出来当作测试集, 循环.  

  
对于小规模数据集而言，这可能不会太耗时，但是对于大规模数据集来说，LOOCV的计算开销会很大。  
另外，LOOCV还有一些其他的缺点。例如，对于大规模数据集，由于每次只使用一个样本作为验证集，模型可能无法充分学习整个数据集的特征，因此可能会导致模型性能的偏差。此外，LOOCV的结果可能会对数据集中的异常值敏感，因为每次只有一个样本被排除，异常值的影响无法被平均化。
